```python
import torch
```

```python
data = torch.load("./models/resnet18/resnet18_59.pth")
loss_history_train = data["loss_history_train"]
loss_history_val = data["loss_history_val"]
```

```python
loss_epoch_train = data["loss_epoch_train"]
loss_epoch_val = data["loss_epoch_val"]
acc_epoch_val_top1 = data["acc_epoch_val_top1"]
acc_epoch_val_top5 = data["acc_epoch_val_top5"]
```

```python
print(len(loss_history_train))
print(len(loss_history_val))
print(len(loss_epoch_train))
print(len(loss_epoch_val))

len_train = len(loss_history_train)
len_val = len(loss_history_val)
len_epoch = len(loss_epoch_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (16,20)


fig,(ax1,ax2,ax3,ax4) = plt.subplots(4,1)
# fig.subplots_adjust(hspace=0.5)

ax1.plot(np.arange(0,len_train,1),loss_history_train)
ax1.set_xlim(0,len_train)
ax1.set_ylim(0,8)
ax1.set_xlabel("iter")
ax1.set_ylabel("loss")

ax2.plot(np.arange(0,len_val,1),loss_history_val)
ax2.set_xlim(0,len_val)
ax2.set_ylim(0,8)
ax2.set_xlabel("iter")
ax2.set_ylabel("loss")


t = np.arange(0,len_epoch,1)
ax3.plot(t,loss_epoch_train,t,loss_epoch_val,marker='.')
ax3.hlines(1.5, 0, 66.0, colors = "g", linestyles = "dashed")
ax3.set_xlim(0,len_epoch)
ax3.set_ylim(0.5,2.0)
ax3.set_xlabel("epoch")
ax3.set_ylabel("loss")


t = np.arange(0,len_epoch,1)
ax4.plot(t,acc_epoch_val_top1,t,acc_epoch_val_top5,marker='.')
ax4.hlines(70, 0, 66.0, colors = "c", linestyles = "dashed")
ax4.hlines(90, 0, 66.0, colors = "r", linestyles = "dashed")
ax4.set_xlim(0,len_epoch)
ax4.set_ylim(65,95)
ax4.set_xlabel("epoch")
ax4.set_ylabel("Prce")


plt.show()
```
